  ## Importing packages and data
 

```python
## Importing all the necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

## Importing the dataframe itself to a new variable, along with displaying it
covid_data = pd.read_csv('https://data.chhs.ca.gov/dataset/e39edc8e-9db1-40a7-9e87-89169401c3f5/resource/de27ce58-edc8-45fb-bebc-08c4b29c5efe/download/covid19postvaxstatewidestats_07172022.csv')
covid_df = covid_data.set_index('date')
covid_df.head(532)
```

  ## Creating Functions

```python
## Made a stats function to return all stats for one column
## Inputs: (data) represent the entire dataset, and (column) refers to the column section from that dataset

def stats_max(data,column):
    '''This function obtains the statistics for a specific column that you ask for'''
    stats_df = data.describe()
    return_column = stats_df.loc[:,column]
    print('Stats for ' + column + ':')
    print(return_column)
```

```python
## This function returns 2 plots in a side by side view
## Inputs: (x) and (y) are the datasets from the columns, (labelx) to (labelT2) are of course meant for the labels in graph

def twoSubplot(x,y,labelx,labely,labelT1,labelT2):
    '''This is for plotting 2 data seperately with labels'''
    #Sets up the plots
    fig, (ax1, ax2) = plt.subplots(1,2,figsize=(15,5))
    ax1.plot(x)
    ax2.plot(y)
    
    #Labels it all
    ax1.set_xlabel(labelx)
    ax2.set_xlabel(labelx)
    ax1.set_ylabel(labely)
    ax2.set_ylabel(labely)
    ax1.set_title(labelT1)
    ax2.set_title(labelT2)
    plt.show()
```

```python
## This function returns 2 datasets into one graph for comparison
## Inputs: (x) and (y) are the datasets from the columns, (labelx) to (labelsecond) are of course meant for the labels in graph

def combinedPlots(x,y,labelx,labely,labelT,labelfirst,labelsecond): 
    '''This is for taking 2 datas and combining it together with labels'''
    #Sets up the plots
    fig, ax = plt.subplots(figsize=(15, 5))
    plt.plot(x)
    plt.plot(y)
    
    #Labels it all
    plt.xlabel(labelx)
    plt.ylabel(labely)
    plt.title(labelT)
    plt.legend([labelfirst,labelsecond])
    plt.show()
```

```python
## This function returns 2 datasets into one graph for comparison while using the ewm method to "smooth" the datasets
## Inputs: (x) and (y) are the datasets from the columns, (labelx) to (labelsecond) are of course meant for the labels in graph

def smoothCombinedPlots(x,y,labelx,labely,labelT,labelfirst,labelsecond):
    '''This function smooths out to show the "trend" of the data'''
    #Sets up the plots
    fig, ax = plt.subplots(figsize=(15, 5))
    ax11 = x.ewm(span = 50).mean()
    ax12 = y.ewm(span = 50).mean()
    ax11.plot()
    ax12.plot()
    
    #Labels it all
    plt.xlabel(labelx)
    plt.ylabel(labely)
    plt.title(labelT)
    plt.legend([labelfirst,labelsecond])
    plt.show()
```

## Unvaccinated vs. Vaccinated Cases

```python
# This section is meant to extract the data out into 2 seperate datasets based on the columns
unv_cases = covid_data['unvaccinated_cases']
v_cases = covid_data['vaccinated_cases']

# This is just for printing out the extracted raw data as a test
print(unv_cases)
print(v_cases)
print("Labeling variables to focus on select columns ")
print('\n')

# This section provides the stats such as the mean, min, and max values of each dataset
stats_max(covid_data,'unvaccinated_cases')
stats_max(covid_data,'vaccinated_cases')
print("This provides descriptive statistical data")
```

```python
# This section is for labeling
laX = 'Number of days'
laY = 'Number of cases'
laT1 = 'Unvaccinated cases'
laT2 = 'Vaccinated cases'
laT3 = 'Unvaccinated and Vaccinated'
laT11 = 'Unvaccinated cases trend'
laT21 = 'Vaccinated cases trend'
laT31 = 'Unvaccinated and Vaccinated trend'

# Presenting visual data
twoSubplot(unv_cases,v_cases,laX,laY,laT1,laT2)
combinedPlots(unv_cases,v_cases,laX,laY,laT3,laT1,laT2)
smoothCombinedPlots(unv_cases,v_cases,laX,laY,laT31,laT11,laT21)

```
